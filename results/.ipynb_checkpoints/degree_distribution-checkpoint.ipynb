# Direct Links and n-degrees

```python
from collections import defaultdict

import pandas as pd
from scipy import stats 
import numpy as np
import json

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#load first link network 
    # runtime < 30min
results_path = "/Users/mark/Desktop/wiki_v4/"
with open(results_path + "fln.json") as f:
    fln = json.load(f)
fldf = pd.DataFrame(fln.items())
fldf.columns = ['article', 'first link']
```

```python
#reverse hash of direct_links
direct_links = defaultdict(list)

for article, first_link in fln.iteritems():
    direct_links[first_link].append(article)
    
with open(results_path + 'direct_links.json', 'w') as fp:
    json.dump(direct_links, fp)
```

```python
dldf = pd.DataFrame(direct_links.items())
```

```python
dldf.columns = ['article', 'direct links']
```

```python
#add ndegree dataframe column
l = lambda x: len(x)
dldf['ndegree'] = dldf['direct links'].map(l)
```

```python
dldf
```

## What are the highest ranking articles by n-degree?
n-degree: number of direct first links to an article

```python
dldf_sorted = dldf.sort(columns='ndegree', ascending=False)
```

# ndegree on log-log scale

```python
dldf_sorted['rank'] = np.arange(1, dldf_sorted.shape[0]+1)
dldf_sorted['log(rank)'] = np.log(dldf_sorted['rank'])
dldf_sorted['log(ndegree)']=  np.log(dldf_sorted['ndegree'])
```

```python
plt.scatter(dldf_sorted['log(rank)'], dldf_sorted['log(ndegree)'], color="#F08080", label='r = -0.98')
plt.title("log log")
plt.xlabel("log(rank)")
plt.ylabel("log(ndegree)")
plt.legend()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(dldf_sorted["log(rank)"], dldf_sorted["log(ndegree)"])
print slope, intercept, r_value, p_value, std_err 
```
