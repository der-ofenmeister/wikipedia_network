#Traversal Visits

measured as the accumulated number of visits after traversing the network using the river model:
 1. begin at every article
 2. first link receives +1 visit
 3. continue to next first link, +1 visit 
     - stop at loop or dead link

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load results into dataframe (~ 1min runtime)

results_path = "/Users/mark/Desktop/wiki_v4/"
with open(results_path + "clicks.json") as f:
    dict = json.load(f)
df = pd.DataFrame(dict.items())
df.columns = ['article', 'traversals']

#graphs path
path = "/Users/mark/Dropbox/Math/Complex_Systems/research/writeup/graphics/"
```

##Top 50 Pages

```python
top_50 = df.sort(columns='traversals', ascending=False).head(50)
```

```python

x = list(top_50["article"])
y = [v / 1000000.0 for v in top_50["traversals"]]

plt.figure(figsize=(16,16))
sns.set_style("whitegrid")

my_palette = sns.color_palette(["#9ecae1"])

bar_plot = sns.barplot(x=x, y=y,
                       palette=my_palette,
                        order=top_50["article"].tolist())
plt.xticks(rotation=90, fontsize=14)
plt.yticks(rotation=90)

plt.ylabel("traversals (in millions)", fontsize=16)
plt.xlabel("")

plt.savefig(path+'top_articles_list.png', format='png', dpi=300, transparent=True, bbox_inches='tight')
```

##What's the distribution of page visits?

```python
toppdf_10k = df.sort(columns='traversals', ascending=False).head(10000)['traversals'] / 1000000
toppdf_100 = df.sort(columns='traversals', ascending=False).head(100)['traversals'] / 1000000
toppdf_1k = df.sort(columns='traversals', ascending=False).head(1000)['traversals'] / 1000000
toppdf_100 = df.sort(columns='traversals', ascending=False).head(100)['traversals'] / 1000000
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.figure(figsize=(10,10))

visits_plot1 = toppdf_10k.plot(kind='hist', title="Top 10,000 Pages")
visits_plot1.set_xlabel("traversal visits (millions)")
visits_plot1.set_ylabel("frequency")
#plt.savefig(path+'top_articles_list.png', format='png', dpi=300)
```

```python
plt.figure(figsize=(10,10))

sns.set_style("darkgrid")
sns.despine(left=True, trim=True)
sns.set_context("paper")
my_palette = sns.color_palette(["#9ecae1"])
sns.set_palette(my_palette)

visits_plot2 = toppdf_1k.plot(kind='hist', title="Top 1,000 Pages", color="y")
visits_plot2.set_xlabel("traversal visits (millions)")
visits_plot2.set_ylabel("frequency")
#plt.savefig(path+'top_1k_articles.png', format='png', dpi=300)
```

```python
sns.distplot(toppdf_1k, color="y")
plt.xlabel("traversal visits (millions)")
plt.ylabel("frequency")
```

```python
visits_plot3 = toppdf_100.plot(kind='hist', title="Top 100 Pages")
visits_plot3.set_xlabel("traversal visits (millions)")
visits_plot3.set_ylabel("frequency")
```
